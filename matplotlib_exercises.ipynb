1. Use matplotlib to plot the following equation:

y = x^2 - x + 2

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.


```python
import matplotlib.pyplot as plt 
print(plt)
```

```python
x = list(range(-50, 50))

def function_for_y(x):
    y = [(x**2) - x + 2 for x in x]
    return y 

y = function_for_y(x)

plt.plot(x, y)
plt.annotate('Origin', xy=(0, 0), xytext=(-5, 400),
arrowprops={'facecolor': 'blue'})
plt.show()
```

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

- You can use functions from the math module to help implement some of the equations above. 



```python
import math 

x = list(range(200))

def square_root_function(x):
    y = [math.sqrt(x) for x in x]
    return y 

y = square_root_function(x)

plt.plot(x, y)
plt.title('y= sqrt of x')
plt.show()
```

```python
x = list(range(200))

def cubed_function(x):
    y = [x * x * x for x in x]
    return y 

y = cubed_function(x)

plt.plot(x, y)
plt.title('y= x^3')
plt.show()
```

```python
x = list(range(200))

def one_over_x_plus_one(x):
    y = [1/(x + 1) for x in x]
    return y 

y = one_over_x_plus_one(x)

plt.plot(x, y)
plt.title('y= 1 / (x + 1)')
plt.show()
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize = (10, 10))

x = list(range(200))
y1 = square_root_function(x)
y2 = cubed_function(x)
y3 = power_of_x_function(x)
y4 = one_over_x_plus_one(x)

plt.suptitle('Figures Subplot', fontsize = 17)

plt.subplot(2,2,1)
plt.plot(x, y1)
plt.title('y= sqrt of x')

plt.subplot(2,2,2)
plt.plot(x, y2)
plt.title('y= x^3')

plt.subplot(2,2,3)
plt.plot(x, y3)
plt.title('y= 2^X')

plt.subplot(2,2,4)
plt.plot(x, y4)
plt.title('y= 1 / (x + 1)')

plt.tight_layout
plt.show()



```
