```python
import matplotlib.pyplot as plt 
import math
import numpy as np
```

```python
plt.figure(figsize = (8, 8))

plt.title('Big O Notation')
plt.xlabel("Elements")
plt.ylabel('Operations')

x = list(range(1, 100))
n = np.array(x)

y1 = 0*n + 1
y2 = np.log(n)
y3 = n
y4 = n * np.log(n)
y5 = n**2
y6 = 2**n
y7 = [math.factorial(n) for n in n]
y8 = n**n

plt.plot(n, y1, label = '$O(1)$')
plt.plot(n, y2, label = 'O(log n)')
plt.plot(n, y3, label = '$O(n)$')
plt.plot(n, y4, label = 'O(n log n)')
plt.plot(n, y5, label = '$O(n^2)$')
plt.plot(n, y6, label = '$O(2^n)$')
plt.plot(n, y7, label = '$O(n!)$')
plt.plot(n, y8, label = '$O(n^n)$')

plt.xlim(0, 15)
plt.ylim(0, 15)

plt.legend(loc= 1)
plt.show()
```
